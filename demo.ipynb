```python


import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from tqdm import tqdm
```

```python
data=[os.path.join('Weed_Detection/data/')]
```

```python
images = [data[x] for x in range(len(data)) if data[x].endswith('.jpeg')]
labels = [data[x] for x in range(len(data)) if data[x].endswith('.txt')]
```

```python
df = pd.DataFrame(columns=['image','class','x_min','y_min','x_max','y_max'])
df
```

```python
i=0
for id in tqdm(images):
    label = open(id[:-5]+'.txt','r') 
    a = label.read()
    label.close()
    label = a.split('\n')
    cleaned_label = label[:-1]
    img = id.split('/')
    for name in cleaned_label:
        ind_label = name.split(' ')
        df.loc[i,'image'] = img[-1]
        df.loc[i,'class'] = int(ind_label[0])
        df.loc[i,'x_min'] = float(ind_label[1])
        df.loc[i,'y_min'] = float(ind_label[2])
        df.loc[i,'x_max'] = float(ind_label[3])
        df.loc[i,'y_max'] = float(ind_label[4])
        i += 1
```

```python
df=df.drop(["y_max",'y_min','x_max','x_min'],axis=1)
df
```

```python
images=[]
labels=[]
for i in range(len(df)): 
    img=cv2.imread('Weed_Detection/data/'+'/'+str(df['image'][i]))
    img=cv2.resize(img,(227,227))
    images.append(img)
    labels.append(df['class'][i])
```

```python
images = np.array(images)
```

```python
y=df['class'].values
Y=y.reshape(-1,1)
Y.shape
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(images, Y, test_size=0.1, random_state=455)
#train_y = train_y.toarray().reshape(-1, 2)
#inpect the shape of the training and testing.
train_y = train_y.astype('float32')
test_y = test_y.astype('float32')
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
import tensorflow as tf

# Normalize the images
x_train = train_x / 255.0
x_test = test_x / 255.0


# Define the model architecture
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(227, 227, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(10, activation='relu'),

    tf.keras.layers.Dense(2, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(x_train, train_y, epochs=10, validation_data=(x_test, test_y))
```

```python
model.save("self_build_02.h5")
```
